# LR实现线性预测
## 任务
基于给定的数据集，实现线性回归模型，预测x=3.5对应的y值，并进行模型评估。

```python
# 数据加载
import pandas as pd
data = pd.read_csv('./generated_data.csv')
data.head()
```

```python
# data赋值
x = data.loc[:,"x"]
y = data.loc[:,"y"]
print(x,y)
```

```python
# 可视化数据
from matplotlib import pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(x,y)
plt.show()
```

```python
#建立模型 set up a linear regression model
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
import numpy as np
# 转换数据
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

```python
# 训练模型
lr_model.fit(x,y)
```

```python
# 预测
y_predict = lr_model.predict(x)
print(y_predict)
print(y)
```

```python
# 预测x = 3.5的y值
y_3 = lr_model.predict([[3.5]])
print(y_3)
```

```python
# a\b打印
a = lr_model.coef_
b = lr_model.intercept_
print(a)
print(b)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
# 计算均方误差
mse = mean_squared_error(y, y_predict)
print("均方误差：", mse)
# 计算R2分数
r2 = r2_score(y, y_predict)
print("R2分数：", r2)
```
