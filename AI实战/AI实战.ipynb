```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,3,4,5,6,7,8,9,10]
print(x,y)
```

```python
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('My first graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
fig2 = plt.figure(figsize=(10, 5))
plt.scatter(x, y)
plt.title('My first scatter plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
